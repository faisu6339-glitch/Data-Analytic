```python
import pandas as pd
```

```python
#Series
```

```python
x=[3,4,5,6,7,8]
var=pd.Series(x)
print(var)
print(type(var))
print(var[2])
```

```python
#Changing Index in Series
```

```python
#List
```

```python
y=[33,445,65,786,712,800]
var1=pd.Series(y,index=['a','s','d','f','g','h'],dtype="float")
print(var1)
print(type(var1))
```

```python
#Dictionary
```

```python
dic={"name":['python','c','java','c++'],"por":[12,13,14,15],"rank":[1,4,3,2]}
var2=pd.Series(dic)
print(var2)
```

```python
#Creating Series with single data
```

```python
s=pd.Series(12)
print(s)
print(type(s))
```

```python
#Creating Series with single value with multiple times
```

```python
q=pd.Series(12,index=[1,2,3,4,5,6,7,8,9])
print(q)
print(type(q))
```

```python
#Series working with MISSING Data
```

```python
q1=pd.Series(12,index=[1,2,3,4,5,6,7,8,9])
q2=pd.Series(12,index=[1,2,3,4])
print(q1+q2)
```
