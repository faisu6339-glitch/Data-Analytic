Accessing elements of Series (Position Based Indexing)

```python
import numpy as np
import pandas as pd
s=pd.Series([1,2,3,4,5])
print("Element at position",s[3])
```

```python
import numpy as np
import pandas as pd
s=pd.Series([11,12,13,14,15])
print("Element at position\n",s[0:])
print("Element at position\n",s[0:4])
print("Element at position\n",s[:5])
```

```python
import pandas as pd
s=pd.Series([11,12,13,14,15])
print(s[[0,2,3]])
```

```python
import pandas as pd
s=pd.Series([11,12,13,14,15])
print("Last Element",s.iloc[-1])
```

```python
import pandas as pd

s = pd.Series([100, 200, 300, 400], index=["a", "b", "c", "d"])

print("Element at index 'b':", s["b"])
print("First element using position:", s.iloc[0])
print("Slice from b to d:")
print(s["b":"d"])  # label based slicing (inclusive)
```

6.1 DataFrame from a Dictionary of Lists

7. Basic Attributes of DataFrame

Student Table

```python
import pandas as pd

# Step 1: Create data (ALL lists must be same length)
names = ["Aaliya", "Shadmani", "Ummehani", "Zainab", "Talat",
         "Noor Ayesha", "Noor Aamna", "Noor Rahmat", "Noor Azmat", "Shaziya", "Shabnam"]

ages = [21, 22, 20, 23, 22, 21, 20, 23, 22, 21, 24]

marks = [85, 90, 88, 75, 80, 89, 77, 84, 91, 73, 88]

# Step 2: Create a DataFrame
df = pd.DataFrame({
    "Name": names,
    "Age": ages,
    "Marks": marks
})

# Step 3: Show DataFrame
print("Full DataFrame:\n")
print(df)

# Step 4: Show basic info
print("\nShape (Rows, Columns):", df.shape)
print("\nColumn Names:", df.columns)
print("\nData Types:\n", df.dtypes)

# Step 5: Access specific column
print("\nNames of students:")
print(df["Name"])

# Step 6: Access a specific row (second student)
print("\nDetails of second student:")
print(df.loc[1])
```
