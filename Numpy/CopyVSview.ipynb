```python
#Copy VS view
```

```python
import numpy as np
```

```python
var=np.array([1,2,3,4])
co=var.copy()
print("var: ",var)
print("copy: ",co)
```

```python
#View
```

```python
var1=np.array([7,8,9,10])
vi=var1.view()
print("var1: ",var1)
print("view: ",vi)
```

```python
#Showing difference
```

```python
var2=np.array([1,2,3,4])
co=var2.copy()

var2[1]=40

print("var: ",var2)
print("copy: ",co)
```

```python
var3=np.array([7,8,9,10])
vi=var3.view()
var3[1]=60
print("var1: ",var3)
print("view: ",vi)
```
