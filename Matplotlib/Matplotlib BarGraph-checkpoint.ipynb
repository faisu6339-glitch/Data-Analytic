```python
import matplotlib.pyplot as plt
```

```python
#Bar & Plot
```

```python
x=[1,2,3,4]
y=[2,4,6,8]

plt.plot(x,y)
plt.show()
```

```python
a=[1,2,3,4]
b=[2,4,6,8]

plt.bar(a,b,color="r")
plt.show()

```

```python
a=[1,2,3,4]
b=[2,4,6,8]
c=["r",'y',"g",'b']
plt.bar(a,b,color=c,linestyle=":",edgecolor="r",linewidth=2.8)
plt.show()
```

```python
#Bar Plot
```

```python
x=["python","c","java","c++"]
y=[85,97,76,68]
plt.xlabel("language",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Aaliya Fatma",fontsize=20)

c=["r",'y',"g",'b']
plt.bar(x,y,color=c,width=0.5,edgecolor="pink",linewidth=4,alpha=0.5,label="Faisal Sheikh")
plt.show
plt.legend()
```

```python
x=["python","c","java","c++"]
y=[85,97,76,68]
z=[75,80,70,60]

plt.xlabel("language",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Aaliya Fatma",fontsize=20)

width=0.4
c=["r",'y',"g",'b']

plt.bar(x,y,width,color="r",label="Faisal Sheikh")
plt.bar(x,z,width,color="pink",label="Aaliya Fatma")
plt.legend()
plt.show()
```

```python
import numpy as np

x=["python","c","java","c++"]
y=[85,97,76,68]
z=[75,80,70,60]

p=np.arange(len(x))
width=0.4
p1=[j+width for j in p]               

plt.xlabel("language",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Aaliya Fatma",fontsize=20)


c=["r",'y',"g",'b']

plt.bar(p,y,width,color="r",label="Faisal Sheikh")
plt.bar(p1,z,width,color="pink",label="Aaliya Fatma")

plt.xticks(p+width,x,rotation=20)
plt.legend()
plt.show()

```

```python
x=["python","c","java","c++"]
y=[85,97,76,68]
z=[75,80,70,60]

plt.xlabel("Numbers",fontsize=15)
plt.ylabel("language",fontsize=15)
plt.title("Aaliya Fatma",fontsize=20)



plt.barh(x,y,color="r",label="Faisal Sheikh")
plt.barh(x,z,color="pink",label="Aaliya Fatma")

plt.legend()
plt.show()
```
