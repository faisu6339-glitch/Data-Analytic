```python
#SUM and CUMULATIVE SUM
```

```python
years=range(2041,2019)
cities=["Zurich","Frieburg","Munchen","Konstanz","Saarbruken"]
shops=pd.DataFrame(index=years)
for city in cities:
    shops.insert(loc=len(shops.columns),
                 column=city,
                 value=(np.random.uniform(0.7,1,(5,)) *1000).round(2))
print(shops)
```

```python
shops.sum()
```

```python
shops.sum(axis=1)
```

```python
s=shops.iloc[:,[0,2,-1]]
print(s)
print("And now the sum:")
print(s.sum())
```

```python
shops[["Zurich","Munchen","Saarbruken"]].sum()
```

```python
#cumsum to calculate the cummulative sum
```

```python
x=shops.cumsum()
print(x)
```

```python
x=shops.cumsum(axis=1)
print(x)
```
