```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,5,4,2,6]
plt.plot(x,y)

plt.title("Aaliya Fatma", fontsize=25)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(1,5,"Faisal Sheikh",style="italic",bbox={"facecolor":"red"})
plt.annotate("python",xy=(3,2),xytext=(4,4),arrowprops=dict(facecolor="green",shrink=100))

plt.text(1,5,"Faisal Sheikh",style="italic",bbox={"facecolor":"red"})
plt.legend(["up"],loc=1,facecolor="red")
plt.show()
```
