```python
#dropna & fillna 
```

```python
import pandas as pd
```

```python
var=pd.read_csv("locations.csv")
var
```

```python
#To remove all the Empty cell along Y-AXIS(ROW)
```

```python
var.dropna(axis=0)
```

```python
#To remove all the Empty cell along X-AXIS(Column)
```

```python
var.dropna(axis=1)
```

```python
#To remove a single empty row
```

```python
var.dropna(how="any")
```

```python
#TO show Na values
```

```python
var.dropna(how="all")
```

```python
#To remove null value in particular ROW
```

```python
var.dropna(subset=["hospital_beds_per_thousand"])
```

```python
#Remove NULL values
```

```python
var.dropna(inplace=True)
var
```

```python
#Multiple NULL values in single column
```

```python
var=pd.read_csv("locations.csv")
var
```

```python
var.dropna(thresh=6)
```

```python
#Filling the NULL VALUES
```

```python
var=pd.read_csv("locations.csv")
var
```
