```python
#Insert
```

```python
import numpy as np
```

```python
#1d-Array
```

```python
var=np.array([1,2,3,4,5,])
print(var)

np.insert(var,0,9)
```

```python
var1=np.array([1,2,3,4,5,6])
print(var1)

np.insert(var,(0,1),55)
```

```python
#2d-Array
```

```python
var2=np.array([[1,2,3],[4,5,6]])
v2=np.insert(var2,0,55,axis=1)
print(v2)
```

```python
#Multiple Data Passing
```

```python
var3=np.array([[1,2,3],[4,5,6]])
v3=np.insert(var3,2,[22,33],axis=1)
print(v3)
```

```python
#Insert data using append function
```

```python
var4=np.array([1,2,3,4,5,])
print(var4)
x=np.append(var4,6.5)
print(x)
```

```python
#Delete
```

```python
var5=np.array([1,2,3,4,5,])
print(var5)
d=np.delete(var5,1)
print(d)
```
