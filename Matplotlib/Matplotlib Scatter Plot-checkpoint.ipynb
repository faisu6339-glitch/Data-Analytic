```python
import matplotlib.pyplot as plt
```

```python
#SCATTER PLOT
```

```python
day=[1,2,3,4,5,6,7]
Num=[2,5,7,6,8,1,3]
plt.scatter(day,Num)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Num",fontsize=15)
plt.show()
```

```python
#Changing Color of Scatter plot
```

```python
day=[1,2,3,4,5,6,7]
Num=[2,5,7,6,8,1,3]
colors=["r","y","g","b","r","g","r"]
sizes=[100,200,400,500,600,400,300]
plt.scatter(day,Num,c=colors,s=sizes,alpha=0.5)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Num",fontsize=15)
plt.show()
```

```python
#Using Markers
```

```python
day=[1,2,3,4,5,6,7]
Num=[2,5,7,6,8,1,3]
colors=["r","y","g","b","r","g","r"]
sizes=[100,200,400,500,600,400,300]

plt.scatter(day,Num,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Num",fontsize=15)
plt.show()
```

```python
#Using COLOR MAP
```

```python
day=[1,2,3,4,5,6,7]
Num=[2,5,7,6,8,1,3]
colors=[10,40,20,30,29,18,60]
sizes=[100,200,400,500,600,400,300]

plt.scatter(day,Num,c=colors,s=sizes,marker="*",cmap="viridis")
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Num",fontsize=15)
plt.show()
```

```python
#Using Multiple Parameters
```

```python
day=[1,2,3,4,5,6,7]
Num1=[2,5,7,6,8,1,3]
Num2=[3,5,6,7,2,8,5]

colors=[10,40,20,30,29,18,60]
sizes=[100,200,400,500,600,400,300]

plt.scatter(day,Num,c=colors,s=sizes,marker="*",cmap="viridis")
plt.scatter(day,Num2,color="r",s=sizes)

t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Num",fontsize=15)
plt.show()
```
