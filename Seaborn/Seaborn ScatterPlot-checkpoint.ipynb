```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)
plt.show()
```

```python
#Knowing sex
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex")
plt.show()
```

```python
var=sns.load_dataset("penguins").head(20)
var
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",
                sizes=(100,10))
plt.show()
```

```python
#Color parameter
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",sizes=(100,60),
                palette="gist_ncar")
plt.show()
```

```python
#alpha parameter
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",sizes=(100,60),
                palette="gist_ncar",alpha=0.5)
plt.show()
```

```python
#marker parameter
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",sizes=(500,600),
                palette="gist_ncar",markers=m)
plt.show()
```
