```python
#1d array
```

```python
import numpy as np
```

```python
var=np.array([9,8,7,6,5,4,3,2,1,0])
print(var)
print()
for i in var:
    print(i)
```

```python
var1=np.array([[1,2,3,4],[1,2,3,4]])
print(var1)
print()
for j in var1:
    print(j)
print()

for k in var1:
    for l in k:
        print(l)
```

```python
#3d Array
```

```python
var3=np.array([[[1,2,3,4],[1,2,3,4]]])
print(var3)
print(var3.ndim)
print()

for i in var3:
    for j in i:
        for k in j:
            print(k)
```

```python
#Using nditer function
```

```python
var4=np.array([[[1,2,3,4],[1,2,3,4]]])
print(var4)
print(var4.ndim)
print()

for i in np.nditer(var4):
    print(i)
```

```python
#using ndenumerate function for index value
```

```python
var5=np.array([[[1,2,3,4],[1,2,3,4]]])
print(var5)
print(var5.ndim)
print()

for i,d in np.ndenumerate(var5):
    print(i,d)
```
