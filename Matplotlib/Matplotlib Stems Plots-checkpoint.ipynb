```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y)
plt.show()
```

```python
#Creating Dotted line
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":")
plt.show()
```

```python
#Using Marker
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="+r")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="ro")
plt.show()
```

```python
#Bottom parameter
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="+r",bottom=1)
plt.show()
```

```python
#Changing base line color
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="ro",basefmt="g")
plt.show()
```

```python
#Using label parameter
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="+r",label="python")
plt.legend()
plt.show()
```

```python
#Changing orientation
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="ro",orientation="horizontal")
plt.show()
```
