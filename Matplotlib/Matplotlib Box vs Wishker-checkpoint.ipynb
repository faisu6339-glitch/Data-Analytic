```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"],showmeans=True)
plt.show()
```

```python
#Creating Outler
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],showmeans=True,sym="r+")
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"],showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="g"),fillerprops=dict)
plt.show()
```
