```python
#Join Array
```

```python
import numpy as np
```

```python
#1D- ARRAY
```

```python
var=np.array([1,2,3,4,])
var1=np.array([5,6,7,8,9])
arr=np.concatenate((var,var1))
print(arr)
```

```python
#2D- ARRAY
```

```python
var2=np.array([[1,2],[3,4,]])
var3=np.array([[5,6,],[8,9]])
arr1=np.concatenate((var2,var3),axis=1)
print(var2)
print()
print(var3)
print()
print(arr1)
```

```python
#Using stack function
```

```python
var4=np.array([1,2,3,4,])
var5=np.array([5,6,8,9])
arr2=np.stack((var4,var5),axis=1)
print(var4)
print()
print(var5)
print()
print(arr2)
```
