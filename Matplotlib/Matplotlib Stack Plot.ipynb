```python
import matplotlib.pyplot as plt
```

```python
# Single Stack Plot
```

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]

plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area=[2,3,5,6,4]

plt.stackplot(x,area)
plt.show()
```

```python
#Multiple Area Plot
```

```python
x=[1,2,3,4,5]
area1=[1,2,3,4,5]
area2=[1,5,3,4,6]
area3=[1,3,2,4,2]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
#Using Labels
```

```python
x=[1,2,3,4,5]
area1=[1,2,3,4,5]
area2=[1,2,3,4,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
#Changing Colors
```

```python
x=[1,2,3,4,5]
area1=[1,2,3,4,5]
area2=[1,2,3,4,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","b"])
plt.legend(loc=2)
plt.show()
```

```python
#Changing Base line
```

```python
x=[1,2,3,4,5]
area1=[1,2,3,4,5]
area2=[1,2,3,4,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","b"],baseline="wiggle")
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[1,2,3,4,5]
area2=[1,2,3,4,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","b"],baseline="zero")
plt.title("Pyh")
plt.legend(loc=2)
plt.show()
```
