```python
#Bar Plot
```

```python
import matplotlib.pyplot as plt
```

```python
x=["Python","C","C++","Java"]
y=[90,50,70,82]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("Bar Plot",fontsize=20)
c=["y","b","m","b","g"]

plt.bar(x,y,width=0.5,color=c,label="Bar Plot")
plt.legend()
plt.show()
```
