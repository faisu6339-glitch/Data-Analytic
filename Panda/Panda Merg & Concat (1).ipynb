```python
#Merge
```

```python
import pandas as pd
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.merge(var1,var2,on="A")
```

```python
#Showing right
```

```python
var3=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var3,var4,how="right")
```

```python
#Using indicator
```

```python
var3=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var3,var4,how="outer",indicator=True)
```

```python
 #Using indexing
```

```python
var3=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.merge(var3,var4,left_index=True,right_index=True)
```

```python
#Changing name of index using suffix
```

```python
var3=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.merge(var3,var4,left_index=True,right_index=True,suffixes=("_name","_id"))

```

```python
#Concat
```

```python
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([11,12,13,14])
pd.concat([sr1,sr2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2])
```

```python
#Using axis
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.concat([d1,d2],axis=1)
```

```python
#Removing some values
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2],axis=1)
```

```python
#Using join parameter to remove NaN
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2],axis=1,join="inner")
```

```python
#Using key parameter
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
pd.concat([d1,d2],axis=0,keys=["d1","d2"])
```

```python
#Adding two different Dataframes
```

```python
d1=pd.DataFrame({"D":[1,2,3,4]})
d2=pd.DataFrame({"A":[1,2],"C":[21,22]})
pd.concat([d1,d2])
```
