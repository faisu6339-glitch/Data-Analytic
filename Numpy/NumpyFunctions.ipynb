```python
import numpy as np
```

```python
#shuffle
```

```python
var =np.array([1,2,3,4,5])
np.random.shuffle(var)
print(var)
```

```python
#unique
```

```python
var1 =np.array([1,2,3,4,5,6,1,1,2,1,2,3,3,4,5])
x=np.unique(var1,return_index=True,return_counts=True)
print(x)
```

```python
#resize
```

```python
var2 =np.array([1,2,3,4,5,6,])
y=np.resize(var2,(2,3))
print(y)
```

```python
#flatten
```

```python
var3 =np.array([11,22,33,44,55,66,])
z=np.resize(var3,(2,3))
print(z)
print()
print(z.flatten())
```
