```python
import matplotlib.pyplot as plt
```

```python
#Step Plot
```

```python
x=[1,2,3,4,5]
y=[2,2,3,5,4]

plt.step(x,y)

plt.title("python")
plt.xlabel("x- axis")
plt.ylabel("y- axis")
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,2,3,5,4]

plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="Python")

plt.title("python")
plt.xlabel("x- axis")
plt.ylabel("y- axis")

plt.legend(loc=2)
plt.grid()
plt.show()
```
