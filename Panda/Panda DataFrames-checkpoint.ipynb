```python
import pandas as pd
```

```python
#DataFrames
```

```python
a=[1,2,3,4,5,6]
var=pd.DataFrame(a)
print(var)
print(type(var))
```

```python
#Creating Data Frames using Dictionary
```

```python
d={"a":[1,2,3,4,5,6],"s":[11,22,33,44,55,66],1:[1,2,3,4,5,6]}
var1=pd.DataFrame(d)
print(var1)
print()

print(type(var1))
print()
#Changing Indexing
var1=pd.DataFrame(d,columns=["a","s",1],index=["a","b","c","d","e","f"])
print(var1)

#To get a particular Data
print(var1["a"][3])
```

```python
#Creating Data Frames using lists
```

```python
list_1=[[1,2,3,4,5],[11,22,33,44,55]]
var2=pd.DataFrame(list_1)
print(var2)
print(type(var2))
```

```python
#Creating Data Frames using Series
```

```python
sr={"s":pd.Series([1,2,3,4]),"r":pd.Series([11,22,33,44])}
var3=pd.DataFrame(sr)
print(var3)
print(type(var3))
```
