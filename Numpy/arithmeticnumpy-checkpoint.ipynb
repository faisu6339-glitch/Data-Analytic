```python
#Arithmetic Operations on 1-D Array
```

```python
import numpy as  np
```

```python
var=np.array([1,2,3,4])
varadd=var+3
print(varadd)
```

```python
var=np.array([1,2,3,4])
varadd=var%3
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([1,2,3,4])
varadd=var1+var2
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([1,2,3,4])
varadd=var1-var2
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,7,3,6])
varadd=var1*var2
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([1,7,8,9])
varadd=var1/var2
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,2,6,4])
varadd=var1%var2
print(varadd)
```

```python
#Using function for arithmetic operations 
```

```python
var1=np.array([1,2,3,4])
var2=np.array([1,7,8,9])
varadd=np.add(var1,var2)
print(varadd)
```

```python
var=np.array([1,2,3,4])
varadd =np.reciprocal(var)
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,2,6,4])
varadd=var1%var2
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,2,6,4])
varadd=np.subtract(var1,var2)
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,2,6,4])
varadd=np.multiply(var1,var2)
print(varadd)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,2,6,4])
varadd=np.power(var1,var2)
print(varadd)
```

```python
#Arithmetic Operations on 2-D Array
```

```python
var21=np.array([[1,2,3,4],[1,2,3,4]])
var22=np.array([[1,2,3,4],[1,2,3,4]])

varadd2 =var21+var22
print(varadd2)
```

```python
var21=np.array([[1,2,3,4],[1,2,3,4]])
var22=np.array([[1,2,3,4],[1,2,3,4]])

print(var21)
print()
print(var22)
print()

varadd2 =var21*var22
print(varadd2)
```
