```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
```

```python
var=sns.load_dataset("penguins")
var
```

```python
#Plotting Bar plot
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var)
plt.show()
```

```python
#hue parameter for separating data
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex")
plt.show()
```

```python
#order parameter for changing graph
```

```python
order_1=["Dream","Torgesen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
            order=order_1,hue_order=["Female","Male"])
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
order=order_1,hue_order=["Female","Male"])
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="v")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="h")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,color="r")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,palette="PRGn")
plt.show()
```

```python
#palette
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
order=order_1,hue_order=["Female","Male"],palette='icefire')
plt.show()
```

```python
#Saturation
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
order=order_1,hue_order=["Female","Male"],saturation=0.3)
plt.show()
```

```python
#errcolor
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
order=order_1,hue_order=["Female","Male"],saturation=10,errcolor="y",errwidth=10)
plt.show()
```

```python
#Using cap
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
order=order_1,hue_order=["Female","Male"],saturation=10,capsize=0.7,errcolor="b")
plt.show()
```
