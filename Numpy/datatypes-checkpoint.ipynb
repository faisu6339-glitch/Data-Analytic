#Data type


```python
import numpy as np
```

```python
#creating an array
```

```python
var=np.array([1,2,3,4,5])
print("Data Type : ",var.dtype)
```

```python
var=np.array([16,32,53,42,15])
print("Data Type : ",var.dtype)
```

```python
var=np.array([1.0,2.1,3.0,4.1,5.0])
print("Data Type : ",var.dtype)
```

```python
var=np.array(["A","B","C","D"])
print("Data Type : ",var.dtype)
```

```python
var=np.array(["A","B","C","D",1,2,3,4])
print("Data Type : ",var.dtype)
```

```python
#Convert Data types into another Datatype
```

```python
x=np.array([1,2,3,4,5])
print("Data  Type : ",x.dtype)
```

```python
x=np.array([1,2,3,4,5] ,dtype=np.int8)
print("Data  Type : ",x.dtype)
print(x)
```

```python
x1=np.array([1,2,3,4,5] ,dtype="f")
print("Data  Type : ",x1.dtype)
print(x1)
```

```python
#use Data Type as a funcion
```

```python
x2=np.array([1,2,3,4,5] )
new=np.float32(x2)
print("Data  Type : ",x2.dtype)
print("Data  Type : ",new.dtype)
print(x2)
print(new)
```
