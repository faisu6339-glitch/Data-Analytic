```python
#Histogram Plot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"])
plt.show()
```

```python
#bins parameter
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
#Creating lihne grph by using kde parameter
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True)
plt.show()
```

```python
#rug parameter
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True)
plt.show()
```

```python
#Color parameter
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="r")
plt.show()
```

```python
#log parameter
```

```python
sns.displot(var["flipper_length_mm"],kde=True,rug=True,color="g",log_scale=True)
plt.show()
```
