```python
#Split Array
```

```python
import numpy as np
```

```python
var=np.array([1,2,3,4,5,6,])
print(var)

arr=np.array_split(var,3)
print()
print(arr)
print(arr[0])
print(type(arr))
```

```python
var1=np.array([[1,2],[3,4],[5,6,]])
print(var1)

arr1=np.array_split(var1,3)
arr2=np.array_split(var1,3,axis=1)
print()
print(arr1)
print(arr2)
```
