```python
#Arithmetics Functions used in Array 
```

```python
import numpy as np
```

```python
var=np.array([1,2,3,4,5,6,7,8,3,2,1])
print("min: ",np.min(var),np.argmin(var))
print("max: ",np.max(var),np.argmax(var))
```

```python
var1=np.array([[2,1,3],[9,5,6]])
print(np.min(var1,axis=0))
```

```python
var3=np.array([2,5,8,98,81,25,64])
print("sqrt: ",np.sqrt(var3))
```

```python
var4=np.array([1,2,3])
print(np.sin(var4))
```

```python
var5=np.array([1,2,3])
print(np.cos(var5))
```

```python
var6=np.array([4,76,98,67,90,57])
print(np.cumsum(var6))
```
