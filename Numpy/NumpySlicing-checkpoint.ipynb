```python
import numpy as np
```

```python
var=np.array([1,2,3,4,5,6,7,8])
            # 0,1,2,3,4,5,6,7 
print(var)
print()
print("2 to 5 : ",var[1:5])

print("2 to End :",var[1:])
print("start to 5 :",var[:5])
print("stop:",var[1:5:2])
```
