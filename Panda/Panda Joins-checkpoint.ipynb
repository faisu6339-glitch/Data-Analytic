```python
import pandas as pd
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2=pd.DataFrame({"C":[10,20,],"D":[11,12]},index=["a","b"])
#Outer
var2.join(var1,how="outer")
#Inner
var1.join(var2,how="inner")
```

```python
var3=pd.DataFrame({"E":[1,2,3,4],"F":[11,12,13,14]},index=["a","b","c","d"])
var4=pd.DataFrame({"G":[10,20,],"F":[11,12]},index=["a","b"])
#Outer
var4.join(var3,how="outer",lsuffix="_12")
```
