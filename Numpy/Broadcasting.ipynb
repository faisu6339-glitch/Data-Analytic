```python
#Broadcasting array in numpy
```

```python
import numpy as np
```

```python
var1=np.array([1,2,3])
print(var1.shape)
print(var1)
print()

var2=np.array([[1,],[2],[3]])
print(var2.shape)
print()
print(var2)
print()

print(var1+var2)
```

```python
x=np.array([[1],[2]])
print(x.shape)

y=np.array([[1,2,3],[1,2,3]])
print(y.shape)

print(x+y)
```
