```python
#Line Plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x_1=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x_1)
plt.show()
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1=sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1)
plt.show()
```

```python
#hue parameter
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex")
plt.show()
```

```python
#size parameter
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",size=30)
plt.show()
```

```python
#style parameter
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",size=30,style="sex")
plt.show()
```

```python
#Changing color by using palette parameter
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent")
plt.show()
```

```python
#Markers parameter
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent"
            ,markers=["o",">"])
plt.show()
```

```python
#Decreasing data values by using head() function
```

```python
y_1=sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent"
            ,markers=["o",">"])
plt.show()
```

```python
#dashes parameter for removing dotted line
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent"
            ,markers=["o",">"],dashes=False)
plt.show()
```

```python
#Removing legend
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent"
            ,markers=["o",">"],legend=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent"
            ,markers=["o",">"])
plt.grid()
plt.title("Python")
plt.show()
```
