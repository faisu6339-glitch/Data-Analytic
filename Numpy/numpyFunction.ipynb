#Zeros

```python
import numpy as np
```

```python
arr_zero=np.zeros(4)
arr_zero1=np.zeros((3,4))

print(arr_zero)
print()
print(arr_zero1)
```

```python
arr_one=np.ones(5)
arr_one1=np.ones((6,4))
print(arr_one)
print(arr_one1)
```
