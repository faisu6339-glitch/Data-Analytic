```python
var2=np.array([[[1,2],[6,7]]])
print(var2)
print(var2.ndim)
print()
print(var2[0,1])
print()
print(var2[0,1,1])
```

```python
#Slicing in Array
```

```python
var3=np.array([1,2,3,4,5,6,7,])
print(var3)
print()
print("2 to 5:",var3[1:5])
print()
print(" 2 to End :", var3[1:])
print()
print("Start to 5 :",var3[:5])
print("stop: ",var3[::])
print("stop: ",var3[::2])
print("stop: ",var3[1:5:2])
```
