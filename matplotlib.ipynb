Grouped Bar Plot (Comparison between 2 sets)

```python
import numpy as np

subjects = ["Math", "Science", "English"]
boys = [80, 70, 85]
girls = [90, 78, 88]

x = np.arange(len(subjects))
width = 0.35

plt.bar(x - width/2, boys, width, label='Boys')
plt.bar(x + width/2, girls, width, label='Girls')

plt.xticks(x, subjects)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Boys vs Girls Marks")
plt.legend()
plt.show()
```

Stacked Bar Plot

```python
plt.bar(subjects, boys, label='Boys')
plt.bar(subjects, girls, bottom=boys, label='Girls')

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Stacked Bar Plot")
plt.legend()
plt.show()
```

STACK PLOT VS AREA PLOT




```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area=[1,2,3,4,5,6]
plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
plt.stackplot(x,area,area1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
area3=[1,2,3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"])
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
area3=[1,2,3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"],baseline="zero")
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
area3=[1,2,3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"],baseline="sym")
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
area3=[1,2,3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"],baseline="wiggle")
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
area3=[1,2,3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"],baseline="sym")
plt.title("faisu")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[6,5,4,3,2,1]
area2=[7,9,8,6,14,5]
area3=[1,2,3,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"],baseline="sym")
plt.legend(loc=3)
plt.grid()
plt.show()
```
